```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data X dan Y
x = np.array([13.39817, 14.26733, 14.17275, 14.62525, 13.27167, 14.91675])
y = np.array([1257.56, 1269.23, 1392.5, 1770.25, 1799.63,1800.6])

# Membangun model regresi linier
model = LinearRegression()

# Melatih model dengan data
model.fit(x.reshape(-1, 1), y)

# Memprediksi nilai Y untuk nilai X baru
y_pred = model.predict(x.reshape(-1, 1))

# Menghitung R-squared
r_squared = model.score(x.reshape(-1, 1), y)

# Mencetak koefisien regresi
print("Koefisien regresi (slope):", model.coef_[0])
print("Koefisien intersep:", model.intercept_)

# Plot data dan garis regresi
plt.scatter(x, y)
plt.plot(x, y_pred, color="red")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# Mencetak nilai R-squared
print("R-squared:", r_squared)
```
